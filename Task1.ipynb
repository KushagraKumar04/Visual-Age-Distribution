```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

df.head()

plt.figure(figsize=(10, 6))
plt.hist(df['x1'], bins=30, edgecolor='k', alpha=0.7)
plt.title('x1 Distribution')
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['x2'], bins=30, edgecolor='k', alpha=0.7)
plt.title('x2 Distribution')
plt.xlabel('x2')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

y_counts = df['y'].value_counts()
plt.figure(figsize=(10, 6))
y_counts.plot(kind='bar', color='skyblue', edgecolor='k')
plt.title('y Composition')
plt.xlabel('y')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
